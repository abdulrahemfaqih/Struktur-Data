```python
# import os
# os.system("cls")


# def createMatrix(matrix):
#     global name
#     name = input('')
#     rows = int(input("Jumlah baris = "))
#     columns = int(input("Jumlah kolom = "))
#     for i in range(rows):
#         matrix.append([])
#     for i in range(rows):
#         for j in range(columns):
#             matrix[i].append(0)
#     jumlah_data = int(input("Jumlah data = "))
#     for i in range(jumlah_data):
#         baris = int(input("Baris ke ? "))
#         kolom = int(input("Kolom ke ? "))
#         elemen = int(input(f"Matrix [{baris}, {kolom}] = "))
#         matrix[baris][kolom] = elemen
#     print()


# def printSparseMatrix(matrix):
#     print(name)
#     for row in matrix:
#         anggota = " ".join([str(elemen) for elemen in row])
#         print("| " + anggota + " |")


# def kaliMatrix(matrix1, matrix2):
#     if len(matrix1[0]) != len(matrix2):
#         print('jumlah kolom ')
#     result = []
#     for rows in range(len(matrix1)):
#         row = []
#         for column in range(len(matrix2[0])):
#             sum = 0
#             for i in range(len(matrix2)):
#                 sum += matrix1[rows][i] * matrix2[i][column]
#             row.append(sum)
#         result.append(row)
#     return result

# def main():
#     matrix1 = []
#     matrix2 = []
#     createMatrix(matrix1)
#     createMatrix(matrix2)
#     printSparseMatrix(matrix1)
#     printSparseMatrix(matrix2)
#     hasil_kali = kaliMatrix(matrix1,matrix2)
#     printSparseMatrix(hasil_kali)
# main()




def createMatrix():
    matrix = []
    rows = int(input("Jumlah baris = "))
    columns = int(input("Jumlah kolom = "))
    for i in range(rows):
        matrix.append([])
    for i in range(rows):
        for j in range(columns):
            matrix[i].append(0)
    amountOfData = int(input("Jumlah data = "))
    for i in range(amountOfData):
        row = int(input("baris ke ? "))
        column = int(input("kolom ke ? "))
        element = int(input(f"Matrix [{row}, {column}] = "))
        matrix[row][column] = element
    print()
    return matrix


def printSparseMatrix(matrix):
    for row in matrix:
        elements = "  ".join([str(element) for element in row])
        print(f"| {elements} |")

def multiplyMatrix(matrix1, matrix2):
    result = []
    if len(matrix1[0]) != len(matrix2):
        print("\njumlah kolom pada matrix1 harus sama dengan jumlah baris pada matrix2 ")
    else:
        for row in range(len(matrix1)):
            resultRow = []
            for column in range(len(matrix2[0])):
                sum = 0
                for i in range(len(matrix2)):
                    sum += matrix1[row][i] * matrix2[i][column]
                resultRow.append(sum)
            result.append(resultRow)
    return result

def main():
    print("matrix - 1 ")
    matrix1 = createMatrix()
    print("matrix - 2 ")
    matrix2 = createMatrix()
    print("matrix 1 = ")
    printSparseMatrix(matrix1)
    print("\nmatrix 2 = ")
    printSparseMatrix(matrix2)
    multiplicationMatrix = multiplyMatrix(matrix1, matrix2)
    print("\nhasil kali = ")
    printSparseMatrix(multiplicationMatrix)
main()
```

```python


def buatMatrixnya():
    matrix = []
    jmlhbaris = int(input('jumlah baris = '))
    jmlhkolom = int(input('jumlah kolom = '))
    for i in range(jmlhbaris):
        matrix.append(([]))
    for i in range(jmlhkolom):
        for i in range(jmlhbaris):
            matrix[i].append(0)
    jumlahdata = int(input('jumlah data = '))
    for i in range(jumlahdata):
        baris = int(input('baris ke? '))
        kolom = int(input('kolom ke? '))
        hasil = int(input(f"matriks [{baris},{kolom}] = "))
        matrix[baris][kolom] = hasil  
    return matrix

def nampilinMatrixnya(matrixyangmauditampilin):
    for baris in matrixyangmauditampilin:
        anggota = "  ".join([str(element) for element in baris])
        print(f"| {anggota} |")

def perkalianmatrixnyaJadidibagibagiginibiargapusing(matriksA,matriksB):
    result=[]
    for x in range (len(matriksA)):
        row =[]
        for y in range (len(matriksB[0])):
            total=0
            for z in range (len(matriksB)):
                total= total + (matriksA[x][z] * matriksB[z][y])
            row.append(total)
        result.append(row)
    return result

# panggil di main programnya
print('matrix1')
matrix1 = buatMatrixnya()
print('matrix2')
matrix2 = buatMatrixnya()
print('matrix1 = ')
nampilinMatrixnya(matrix1)
print('matrix2 = ')
nampilinMatrixnya(matrix2)
print('kali = ')
kali = perkalianmatrixnyaJadidibagibagiginibiargapusing(matrix1,matrix2)
nampilinMatrixnya(kali)



    
    

```
